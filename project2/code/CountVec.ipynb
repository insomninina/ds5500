```python
import pandas as pd 
df=pd.read_json('/content/drive/MyDrive/Amazon Reviews/reviews_Beauty_10.json.gz',lines=True,compression='gzip')
```

```python
df1=df[['reviewerID','summary','reviewText','overall']]
```

```python
df1['sentiment'] = np.where(df1['overall']>3, 'pos', 'neg')
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"www.\S+", "", text)
    text_links_removed = "".join([char for char in text if char not in string.punctuation])
    text_cleaned = " ".join([word for word in re.split('\W+', text_links_removed)
        if word not in stopword])
   
    text = " ".join([wn.lemmatize(word) for word in re.split('\W+', text_cleaned)])

    return text
```

```python
stopword = nltk.corpus.stopwords.words('english')
wn = nltk.WordNetLemmatizer()
ps = nltk.PorterStemmer()
words = set(nltk.corpus.words.words())
```

```python
df1['reviewText_clean'] = df1['reviewText'].apply(clean_text)
```

```python
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df1['reviewText_clean'], df1['sentiment'], test_size=0.20,
                                                    random_state = 40)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

from sklearn.linear_model import LogisticRegression
```

```python
vectorizer = CountVectorizer(min_df=4)
```

```python
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
lr_prediction = logreg.predict(X_test)
```

```python
score = logreg.score(X_test, y_test)
print(score)
```

```python
from sklearn.ensemble import RandomForestClassifier
# creating a RF classifier
clf = RandomForestClassifier(n_estimators = 100)  
  
# Training the model on the training dataset
# fit function is used to train the model using the training sets as parameters
clf.fit(X_train, y_train)
  
# performing predictions on the test dataset
y_pred = clf.predict(X_test)
  
# metrics are used to find accuracy or error
from sklearn import metrics  
print()
  
# using metrics module for accuracy calculation
print("ACCURACY OF THE MODEL: ", metrics.accuracy_score(y_test, y_pred))
```
